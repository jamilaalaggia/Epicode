```python
import pandas as pd

#scaricare il dataset in formato CSV oppure in formato JSON

dataset = pd.read_csv ("https://covid.ourworldindata.org/data/owid-covid-data.csv", low_memory = False)
```

```python
#Si richiede di verificare le dimensioni del dataset e le diciture presenti nell'intestazione.
numero_righe = dataset.shape[0]
numero_colonne = dataset.shape[1]
print("Il dataset ha", numero_righe, "righe e", numero_colonne, "colonne.")
```

```python
#Si chiede per ogni continente di trovare il numero di casi totali avvenuti
continenti = dataset[["continent","new_cases"]].groupby("continent").sum()
continenti
```

```python
'''Sempre riguardo i casi di COVID totali, si chiede di sviluppare una funzione che prenda in input 
il dataset e due nomi di
continenti, e che ne confronti i seguenti relativi descrittori statistici: valori minimo e massimo, 
media, e percentuale rispetto al
numero dei casi totali nel mondo (in questo caso calcolati anche sulle locazioni senza indicazione 
di continente).'''

casi_COVID = int(continenti.new_cases.sum())
print("Ci sono stati",casi_COVID, "casi nel mondo")
```

```python
def statistiche_casi(df,cont_a, cont_b):   
    conts = [cont_a, cont_b]
    risultato = []
    for i in conts:
        casi_COVID = int(df.new_cases.sum())
        dati = dataset.loc[:,["continent", "new_cases"]]
        criterio= dataset.continent == i
        dati = dataset.loc[criterio][["continent", "new_cases"]]
        stats = dati.groupby("continent").aggregate(["min", "max", "mean", "sum"])
        stats[("new_cases",  "sum")] *= 100/casi_COVID 
        stats.columns = stats.columns.droplevel(0)
        stats.rename(columns = {'sum':'of_total'}, inplace = True)
        risultato.append(stats) 
    print(casi_COVID)
    return pd.concat(risultato)
```

```python
pd.concat([statistiche_casi(dataset, "South America", "Oceania"), statistiche_casi(dataset, "Europe", "Asia")]).head(3)
```

```python
def statistiche_vaccini(df,cont_a, cont_b):   
    conts = [cont_a, cont_b]
    risultato = []
    for i in conts:
        vaccini = int(df.new_vaccinations.sum())
        dati = dataset.loc[:,["continent", "new_vaccinations"]]
        criterio= dataset.continent == i
        dati = dataset.loc[criterio][["continent", "new_vaccinations"]]
        stats = dati.groupby("continent").aggregate(["min", "max", "mean", "sum"])
        stats[("new_vaccinations",  "sum")] /= vaccini
        stats[("new_vaccinations",  "sum")] *= 100
        stats.columns = stats.columns.droplevel(0)
        stats.rename(columns = {'sum':'of_total'}, inplace = True)
        risultato.append(stats)   
    return pd.concat(risultato)
```

```python
pd.concat([statistiche_vaccini(dataset, "South America", "Oceania"), statistiche_vaccini(dataset, "Europe", "Asia")])
```

In Europa, le statistiche mostrano un alto numero di casi di COVID-19, ma con un tasso di vaccinazioni significativo. Nel Sud America, le infezioni sono state più diffuse, con una minore percentuale di vaccinazioni rispetto all'Europa. D'altra parte, l'Oceania ha registrato il minor numero di casi e una buona percentuale di vaccinazioni.
